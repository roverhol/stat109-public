## Exercise 6 — Two numerical groups: state $H_0$ for `t.test`

We simulate group A from $N(12, 3)$ and group B from $N(12, 3)$ (**same** population mean in both groups). **In words and symbols**, what is the null hypothesis tested by `t.test(y ~ group, var.equal = FALSE)`?

````r
### Solution 6 — code confirms same means in simulation; test still random
set.seed(6)
n_per <- 30
d <- data.frame(
  group = rep(c("A", "B"), each = n_per),
  y = c(rnorm(n_per, 12, 3), rnorm(n_per, 12, 3))
)
t.test(y ~ group, data = d, var.equal = FALSE)
````

### Solution 6 (notation)
- **Symbols:** $H_0: \mu_A = \mu_B$ (sometimes written $\mu_1 = \mu_2$).
- **Words:** The population mean of $y$ in group A equals the population mean of $y$ in group B.
- Here we built the world with both means **12**, so $H_0$ is **true** for the population; the *p*-value only says how surprising the **sample** split would be if that were true.

## Exercise 7 — Two binary groups: state $H_0$ for `prop.test`

Below, both groups have the **same** probability of success **0.25** in the simulation. **State $H_0$** for comparing the two population proportions.

````r
### Solution 7
set.seed(7)
n_per <- 100
pA <- 0.25
pB <- 0.25
d <- data.frame(
  group = rep(c("A", "B"), each = n_per),
  y = c(rbinom(n_per, 1, pA), rbinom(n_per, 1, pB))
)
xa <- sum(d$y[d$group == "A"])
na <- sum(d$group == "A")
xb <- sum(d$y[d$group == "B"])
nb <- sum(d$group == "B")
prop.test(x = c(xa, xb), n = c(na, nb), correct = FALSE)
````

### Solution 7 (notation)
- **Symbols:** $H_0: p_A = p_B$ (same population proportion of successes in A and B).
- **Words:** The probability a randomly chosen unit in group A is a “success” equals that probability in group B.
- We simulated **pA = pB = 0.25**, so $H_0$ is true at the population level.

## Exercise 8 — What does the *p*-value from `t.test` address?

In **one or two sentences**, say what question a *p*-value from `t.test(y ~ group)` answers **without** claiming that a small *p*-value proves your favorite group “causes” larger $y$.

### Solution 8 (example wording)
The *p*-value summarizes how compatible the **observed difference in sample means** is with **“no difference in population means”** ($H_0: \mu_A = \mu_B$) under the test’s assumptions, under the idea that only **random sampling** made the sample means apart. It does **not** by itself prove a causal explanation for why groups differ.

## Exercise 9 — Story with a confounder (numerical **and** binary outcomes)

**Story (fiction).** *GreenFuel* marketed a lunch program (`program = yes/no`). Uptake was higher where cafeteria prices were **low** (`prices = "low"` vs `"high"`). In our universe:

- **Numerical outcome** `score` (weekly well-being): mean depends **only** on `prices`, **not** on `program`.
- **Binary outcome** `improved` (self-report “health improved yes/no”): probability of “yes” depends **only** on `prices`, **not** on `program`.

So there is **no direct causal effect** of `program` on either outcome in the simulation.

**Tasks**

1. Simulate at least **400** people with columns `prices`, `program`, `score`, and `improved` following that story.
2. Report **crude** mean `score` by `program`, then **stratified** means by `program` within `prices`.
3. For the **binary** outcome, run a **crude** `prop.test` comparing `improved` between program yes vs no (build success counts per group), and **stratified** 2×2 counts (or proportions) within each `prices` level.
4. In one sentence each: can a crude *p*-value from `t.test` or `prop.test` **alone** prove the program **caused** better outcomes here?

**Then** run the Solution cell and compare your answers.

````r
### Solution 9 — confounding: program looks helpful until you stratify
set.seed(9)
n <- 500

prices <- sample(c("low", "high"), n, replace = TRUE, prob = c(0.45, 0.55))
p_prog <- ifelse(prices == "low", 0.55, 0.18)
program <- ifelse(rbinom(n, 1, prob = p_prog) == 1, "yes", "no")
mu_score <- ifelse(prices == "low", 72, 58)
score <- rnorm(n, mean = mu_score, sd = 7)
p_improve <- ifelse(prices == "low", 0.62, 0.35)
improved <- rbinom(n, 1, prob = p_improve)

d9 <- data.frame(
  prices = factor(prices),
  program = factor(program),
  score = score,
  improved = improved
)

cat("--- Crude mean score by program ---\n")
print(aggregate(score ~ program, data = d9, FUN = mean))

cat("\n--- Stratified mean score by program within prices ---\n")
print(aggregate(score ~ program + prices, data = d9, FUN = mean))

cat("\n--- Crude t.test on score (NOT causal evidence here) ---\n")
print(t.test(score ~ program, data = d9, var.equal = FALSE))

cat("\n--- Crude prop.test on improved (yes = 1) ---\n")
py <- sum(d9$improved[d9$program == "yes"] == 1)
ny <- sum(d9$program == "yes")
pn <- sum(d9$improved[d9$program == "no"] == 1)
nn <- sum(d9$program == "no")
print(prop.test(x = c(py, pn), n = c(ny, nn), correct = FALSE))

cat("\n--- Stratified: improved counts by program within prices=low ---\n")
dlow <- subset(d9, prices == "low")
py <- sum(dlow$improved[dlow$program == "yes"] == 1)
ny <- sum(dlow$program == "yes")
pn <- sum(dlow$improved[dlow$program == "no"] == 1)
nn <- sum(dlow$program == "no")
print(c(yes_successes = py, yes_n = ny, no_successes = pn, no_n = nn))

cat("\n--- Stratified: improved counts by program within prices=high ---\n")
dhi <- subset(d9, prices == "high")
py <- sum(dhi$improved[dhi$program == "yes"] == 1)
ny <- sum(dhi$program == "yes")
pn <- sum(dhi$improved[dhi$program == "no"] == 1)
nn <- sum(dhi$program == "no")
print(c(yes_successes = py, yes_n = ny, no_successes = pn, no_n = nn))

cat("\nNotes:\n")
cat("- score and improved depend only on prices in the simulation.\n")
cat("- program uptake depends on prices, so crude associations confound prices with program.\n")
````

### Solution 9 (interpretation)
- **Crude** summaries usually make `program = yes` look better on both `score` and `improved`, because more "yes" people come from **low-price** campuses where outcomes are better anyway.
- **Within** each `prices` level, means and improvement rates for yes vs no should be **similar** (only simulation noise), matching **no effect of program** in the data-generating code.
- A small crude *p*-value from `t.test` or `prop.test` does **not** prove the program **caused** better outcomes; the stratified summaries show the **confounding** pattern.
