## Part 4 — For loops (repeat a process and store results)

### 4.0 Anatomy of a `for` loop (what each piece means)

A `for` loop repeats the same steps many times. Here is the basic “shape”:



```r
for (i in 1:n) {
  # code to repeat goes here
}
```


What the parts mean:

- `for (...) { ... }`: tells R “repeat what’s inside the braces `{}`”.
- `i`: the **index variable** (a name you choose). It takes on values one-at-a-time.
- `in 1:n`: the **sequence** of values `i` will go through (here: 1, 2, 3, ..., n).
- `{` and `}`: the braces mark the **body** of the loop (the code that repeats).

Most of the time we also:

- **Pre-allocate storage** (make an empty vector first) so we have a place to save results.
- **Store one result each time** using indexing like `results[i] <- ...`.

#### Making “empty placeholders” (pre-allocation)

When you know you will store **n numbers**, it helps to create the container first:

- `numeric(n)` makes a vector of length `n` filled with zeros (a placeholder for numbers).
- `logical(n)` makes a vector of length `n` filled with `FALSE`.
- `character(n)` makes a vector of length `n` filled with `""` (empty strings).

For matrices, you can do the same idea in 2D:

- `matrix(NA, nrow = r, ncol = c)` makes an `r` by `c` placeholder matrix.

## Part 5 — Probability and simulation (equally likely sample spaces)

Key simulation idea for easy grading:

- Make an **event** into a TRUE/FALSE statement (e.g., “roll is even”).
- The **estimated probability** is just the **proportion of TRUE values**:
