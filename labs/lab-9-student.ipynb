## Question 1 (Numerical variable)

Use `diamonds$carat` as a numerical variable.

- Check its type.
- Compute mean and SD.
- Make a histogram.

### Solution code

```r
class(diamonds$carat)

diamonds %>%
  summarise(
    mean_carat = mean(carat),
    sd_carat = sd(carat)
  )

ggplot(diamonds, aes(x = carat)) +
  geom_histogram(binwidth = 0.1, color = "white") +
  labs(title = "Distribution of carat", x = "Carat", y = "Count")
```

## Question 2 (Categorical variable)

Use `diamonds$cut` as a categorical variable.

- Check its type.
- List category levels.
- Compute count and proportion for each level.

### Solution code

```r
class(diamonds$cut)
levels(diamonds$cut)

diamonds %>%
  count(cut, name = "count") %>%
  mutate(proportion = count / sum(count))
```

## Question 3 (Change variable type)

Treat `color` as numeric-like category codes.

- Create a new variable `color_code` from `color` converted to integer codes.
- Convert `color_code` back to a factor in a new data frame.
- List the levels of the recovered factor.

### Solution code

```r
diamonds_codes <- diamonds %>%
  mutate(color_code = as.integer(color))

glimpse(diamonds_codes %>% select(color, color_code))

diamonds_codes2 <- diamonds_codes %>%
  mutate(color_code = factor(color_code,
                             levels = 1:7,
                             labels = levels(diamonds$color)))

class(diamonds_codes2$color_code)
levels(diamonds_codes2$color_code)
```

## Question 4 (`filter()` with conditions)

Create a subset of diamonds that are:

- `carat >= 1.0`
- `cut` is `"Premium"` or `"Ideal"`

Show the first 10 rows of `carat`, `cut`, `color`, `price`.

### Solution code

```r
diamonds %>%
  filter(carat >= 1.0, cut %in% c("Premium", "Ideal")) %>%
  select(carat, cut, color, price) %>%
  head(10)
```

## Question 5 (`group_by()` + `summarise()`)

Compute grouped summaries for `price` by `cut`:

- `n`
- mean price
- SD price

Then sort by highest mean price.

### Solution code

```r
diamonds %>%
  group_by(cut) %>%
  summarise(
    n = n(),
    mean_price = mean(price),
    sd_price = sd(price)
  ) %>%
  arrange(desc(mean_price))
```
