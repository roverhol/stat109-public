## Part 6 — Simulation: 4 babies returned to 4 mothers

### Story
There are 4 mothers and 4 babies. Babies are returned **at random**.  
We want to study how often matches happen, especially:
- **All 4 correct**  
- **None correct**

### Important idea
This code is written to **mimic the physical card shuffle**:
1. Line up mothers 1–4  
2. Shuffle babies 1–4  
3. Compare positions  
4. Count correct matches  
5. Repeat many times  
6. Use proportions as probability estimates


### 6.1 One trial (one shuffle)
Run this code several times and watch the output change.


````r
mothers <- c(1, 2, 3, 4)

babies <- sample(c(1, 2, 3, 4))   # shuffle the babies
babies

num_matches <- sum(babies == mothers)  # count matches
num_matches
````

### 6.2 Ten trials using a for loop
We will store the number of matches from each trial in a vector.


````r
set.seed(109)  # makes results reproducible for grading

mothers <- c(1, 2, 3, 4)
num_correct <- numeric(10)  # storage (10 empty slots)

for (i in 1:10) {
  babies <- sample(c(1, 2, 3, 4))           # shuffle
  num_correct[i] <- sum(babies == mothers)  # count + record
}

num_correct
````

### 6.3 Many trials (10,000) to estimate probabilities
Now we repeat the shuffle many times and estimate probabilities by proportions.


````r
set.seed(109)

mothers <- c(1, 2, 3, 4)
num_correct <- numeric(10000)

for (i in 1:10000) {
  babies <- sample(c(1, 2, 3, 4))
  num_correct[i] <- sum(babies == mothers)
}

# Estimated probabilities (by simulation)
p_all4  <- mean(num_correct == 4)
p_none  <- mean(num_correct == 0)

p_all4
p_none
````

### 6.4 Distribution of number of matches (optional summary)
This shows how often we got 0 matches, 1 match, 2 matches, etc.


````r
table(num_correct)
````
