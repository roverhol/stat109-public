## Exercise: Correlation, simple linear regression, and interval predictions

Use the dataset already loaded in `dat` with:

- explanatory variable: `Girth` (inches)
- response variable: `Volume` (cubic feet)

**Question:**

1. Compute Pearson correlation between `Girth` and `Volume`.
2. Fit a simple linear regression model `Volume ~ Girth`.
3. For two girth values (`Girth = 10` and `Girth = 16` inches), report:
   - point prediction for `Volume`,
   - 95% confidence interval for the **mean** `Volume` at that girth,
   - 95% **prediction** interval for an **individual** tree’s `Volume` at that girth.

**Desired output:** correlation value, `summary(fit)`, `coef(fit)`, and `predict()` output for both interval types at both `x` values.

```r
# Sample solution (for student handout, replace this cell with blanks / comments only)

# 1) Pearson correlation between girth and volume
cor(dat$Girth, dat$Volume)

# 2) Simple linear regression: Volume ~ Girth
fit <- lm(Volume ~ Girth, data = dat)
summary(fit)
coef(fit)

# 3) Predictions at Girth = 10 and 16 inches (within typical range of these data)
new_x <- tibble(Girth = c(10, 16))

# Point predictions for Volume
predict(fit, newdata = new_x)

# 95% confidence interval for mean Volume at each girth
predict(fit, newdata = new_x, interval = "confidence", level = 0.95)

# 95% prediction interval for Volume of an individual tree at each girth
predict(fit, newdata = new_x, interval = "prediction", level = 0.95)
```
