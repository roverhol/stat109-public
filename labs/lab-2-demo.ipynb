## Part 1 — Data types (what kind of thing is this?)

In R, every value has a **type**. You don’t have to memorize these right away—just know how to *check* the type when you’re unsure.

-   **numeric**: numbers that can include decimals (e.g. `3.14`)
-   **integer**: whole numbers (e.g. `3L` — the `L` means “integer”)
-   **character**: text/strings (e.g. `"STAT 109"`)

Two helpful functions:

- `class(x)` tells you the “kind of thing” R thinks `x` is.

## Part 4 — For loops (repeat a process and store results)

### 4.0 Anatomy of a `for` loop (what each piece means)

A `for` loop repeats the same steps many times. Here is the basic “shape”:



```r
for (i in 1:n) {
  # code to repeat goes here
}
```


What the parts mean:

- `for (...) { ... }`: tells R “repeat what’s inside the braces `{}`”.
- `i`: the **index variable** (a name you choose). It takes on values one-at-a-time.
- `in 1:n`: the **sequence** of values `i` will go through (here: 1, 2, 3, ..., n).
- `{` and `}`: the braces mark the **body** of the loop (the code that repeats).

Most of the time we also:

- **Pre-allocate storage** (make an empty vector first) so we have a place to save results.
- **Store one result each time** using indexing like `results[i] <- ...`.

#### Making “empty placeholders” (pre-allocation)

When you know you will store **n numbers**, it helps to create the container first:

- `numeric(n)` makes a vector of length `n` filled with zeros (a placeholder for numbers).
- `logical(n)` makes a vector of length `n` filled with `FALSE`.
- `character(n)` makes a vector of length `n` filled with `""` (empty strings).

For matrices, you can do the same idea in 2D:

- `matrix(NA, nrow = r, ncol = c)` makes an `r` by `c` placeholder matrix.

### 4.1 A simple for loop over 1 to 5

**Goal output**



```r
[1]  1  4  9 16 25
```


**Your task:** Use a `for` loop to fill a vector with the squares of 1 through 5.



```r
# Solution
results <- numeric(5) # storage: 5 empty slots
for (i in 1:5) {
  # i takes values 1, 2, 3, 4, 5
  results[i] <- i^2
}
results
```

## Part 5 — Probability and simulation (equally likely sample spaces)

Key simulation idea for easy grading:

- Make an **event** into a TRUE/FALSE statement (e.g., “roll is even”).
- The **estimated probability** is just the **proportion of TRUE values**:
  - in R: `mean(event_is_true)` works because `TRUE` counts as 1 and `FALSE` counts as 0.

### 5.1 Equally likely outcomes: one fair die

Let $S = \{1,2,3,4,5,6\}$ be the sample space for a fair die roll (all outcomes equally likely).

**New operator:** `%in%` means “is this value in this set/vector?”

- `x == 2` checks “is x equal to 2?”
- `x %in% c(2, 4, 6)` checks “is x one of 2, 4, or 6?”

For vectors, `%in%` works element-by-element and returns a TRUE/FALSE vector (just like `==` did in Lab 1).

**Your task:**
- Simulate 10,000 fair die rolls.
- Estimate $P(\text{roll is even})$.



```r
# Solution
set.seed(109)

n <- 10000
rolls <- sample(1:6, size = n, replace = TRUE)
# replace = TRUE means outcomes can repeat (like real die rolls)

p_even <- mean(rolls %in% c(2, 4, 6))
# %in% checks membership (is each roll in the set {2,4,6}?)
p_even
```


**Check (theory):** The even outcomes are {2,4,6}, so $P(\text{even}) = 3/6 = 0.5$.

------------------------------------------------------------------------

### 5.2 Two fair dice: probability the sum is 7

**Your task:**
- Simulate 50,000 pairs of fair die rolls.
- Estimate $P(\text{sum} = 7)$.



```r
# Solution
set.seed(109)

n <- 50000
# sample(..., size = n, replace = TRUE) gives n independent rolls
d1 <- sample(1:6, size = n, replace = TRUE)
d2 <- sample(1:6, size = n, replace = TRUE)

sums <- d1 + d2
p_sum7 <- mean(sums == 7)
p_sum7
```


**Check (theory idea):** There are 36 equally likely ordered pairs $(d1,d2)$. Six of them sum to 7: (1,6), (2,5), (3,4), (4,3), (5,2), (6,1), so $P(\text{sum}=7) = 6/36 = 1/6$.

------------------------------------------------------------------------

### 5.3 Using a for loop to simulate many trials

**Your task:** Repeat the “sum of two dice” experiment using a `for` loop that stores each sum in a vector, then estimate $P(\text{sum}=7)$ using `mean()`.



```r
# Solution
set.seed(109)

n <- 20000
sums <- numeric(n)

for (i in 1:n) {
  d1 <- sample(1:6, size = 1)
  d2 <- sample(1:6, size = 1)
  sums[i] <- d1 + d2
}

mean(sums == 7)
```


------------------------------------------------------------------------

### 5.4 (Optional) Build the equally likely sample space explicitly

**Your task:** Use `expand.grid()` to build the 36 equally likely outcomes for two dice, then compute the exact probability that the sum is 7 by counting.



```r
# Solution
space <- expand.grid(d1 = 1:6, d2 = 1:6)
mean(space$d1 + space$d2 == 7)
```
