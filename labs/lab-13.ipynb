## 1) Google Sheets setup (2026 workflow)

Create a sheet with a header row, for example:

- `predictor`
- `response`

Then:

1.  Open your Google Sheet.
2.  Click **Share** -\> set access to **Anyone with the link (Viewer)**
    for class/demo use.
3.  Copy the spreadsheet ID from the URL:
    - URL looks like
      `https://docs.google.com/spreadsheets/d/SPREADSHEET_ID/edit#gid=0`
4.  Build a CSV export URL:
    - `https://docs.google.com/spreadsheets/d/SPREADSHEET_ID/export?format=csv&gid=0`
5.  Use that URL directly in `read_csv()`.

This is the same idea as reading a raw CSV from GitHub/TidyTuesday:
`read_csv("https://...csv")`.

## 2) Demo code (simple and tidyverse-style)

`````r
if (!requireNamespace("tidyverse", quietly = TRUE)) {
  install.packages("tidyverse", repos = "https://cloud.r-project.org")
}
library(tidyverse)
`````

Replace `sheet_csv_url` with your own export link:

`````r
# Example placeholder: replace with your actual sheet CSV export URL
sheet_csv_url <- "https://docs.google.com/spreadsheets/d/YOUR_SHEET_ID/export?format=csv&gid=0"

# dat <- read_csv(sheet_csv_url, show_col_types = FALSE)

# For this page preview only, we create a fallback demo dataset:
set.seed(1313)
dat <- tibble(
  predictor = runif(40, 2, 12),
  response = 8 + 2.5 * predictor + rnorm(40, 0, 3)
)

glimpse(dat)
head(dat, 8)
`````

Fit simple linear model:

`````r
fit <- lm(response ~ predictor, data = dat)
summary(fit)
coef(fit)
`````

Prediction at a new value:

`````r
new_x <- tibble(predictor = 8)

# Point prediction
predict(fit, newdata = new_x)

# Confidence interval for mean response at predictor = 8
predict(fit, newdata = new_x, interval = "confidence", level = 0.95)
`````

Scatterplot with fitted line:

`````r
ggplot(dat, aes(x = predictor, y = response)) +
  geom_point(alpha = 0.8, color = "steelblue4") +
  geom_smooth(method = "lm", se = FALSE, color = "darkred") +
  labs(
    title = "Project 5 demo: response vs predictor",
    x = "Predictor",
    y = "Response"
  ) +
  theme_minimal()
`````

## 4) Common troubleshooting

- **403/permission error:** sheet is not shared for viewing.
- **Wrong sheet tab:** `gid` in URL points to a different tab.
- **Column name mismatch:** check exact header spelling with
  `names(dat)`.
- **Non-numeric columns:** convert with `as.numeric()` only after
  checking raw values/units.
