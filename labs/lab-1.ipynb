## Part 6: Simulation (4 babies returned to 4 mothers)

### Story

There are 4 mothers and 4 babies. Babies are returned at random.

Questions of interest:

- probability all 4 are correctly matched,
- probability none are correctly matched.

### 6.1 One trial (one shuffle)

Run multiple times and observe changes:

```r
mothers <- c(1, 2, 3, 4)

babies <- sample(c(1, 2, 3, 4))
babies

num_matches <- sum(babies == mothers)
num_matches
```

### 6.2 Ten trials with a for loop

```r
set.seed(109)

mothers <- c(1, 2, 3, 4)
num_correct <- numeric(10)

for (i in 1:10) {
  babies <- sample(c(1, 2, 3, 4))
  num_correct[i] <- sum(babies == mothers)
}

num_correct
```

### 6.3 Many trials (10,000) for probability estimates

```r
set.seed(109)

mothers <- c(1, 2, 3, 4)
num_correct <- numeric(10000)

for (i in 1:10000) {
  babies <- sample(c(1, 2, 3, 4))
  num_correct[i] <- sum(babies == mothers)
}

p_all4 <- mean(num_correct == 4)
p_none <- mean(num_correct == 0)

p_all4
p_none
```

### 6.4 Distribution of number of matches

```r
table(num_correct)
```
